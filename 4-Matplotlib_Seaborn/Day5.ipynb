```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Graph')
```

```python
plt.savefig('Test.png')
```

```python
y=x**2
```

```python
plt.plot(x,y,"bo--")
```

##### Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

##### Other examples

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.title("Sine wave")
plt.plot(x,y,'y')
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r')
plt.title('Sine wave')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g')
plt.title('Cosine wave')
```

##### Bar plot

```python
x = [2,8,10]
y= [11,16,9]

x1 = [3,9,11]
y1 = [6,15,8]
```

```python
plt.bar(x,y,align='center')
plt.bar(x1,y1,color="y")
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

##### Histograms

X axis will have the values & Y axis will result the count/density of the values

```python
a = np.array([22,69,5,7,36,77,45,87,99,21,18,91,58])
```

```python
plt.hist(a)
plt.title('Histogram')
```

```python
plt.hist(a,bins=20,color='#FF5733')
```

##### Box Plot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
```

```python
data
```

```python
type(data)
```

```python
len(data)
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
```

##### Pie Chart

```python
labels = 'Batman', 'STR', 'Ramanagawd', 'Salaar'
sizes = [315, 230, 247, 8]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0.1, 0.4)
```

```python
type(labels)
```

```python
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True)
plt.axis('equal')
```

### Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

#### Distribution Plots

distplot

jointplot

pairplot

```python
df.dtypes
```

##### Correlation

```python
df.select_dtypes(include=['float64','int64']).corr()
```

```python
sns.heatmap(df.select_dtypes(include=['number']).corr())
```

##### Joint Plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

##### Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

```python
df['sex'].value_counts()
```

##### Dist plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

##### Categorical Plots

Box plot

Violin plot

Count plot

Bar plot

```python
sns.countplot(data=df)
```

```python
sns.countplot(data=df,x='sex')
```

```python
sns.countplot(y='day',data=df)
```

##### Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(y='total_bill',data=df)
```

##### Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

##### Violin plot

Mix of Box plot & distribution of data in terms of Kernel Density Estimation (KDE)

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
